```python
PTM_graph = nx.Graph()
role_map = {'Source': 0, 'Target': 1}
for idx, row in cleaned_PTM_rows.iterrows():
    PTM_graph.add_nodes_from([
        (row['Official Symbol A'], {"x": torch.tensor([role_map["Source"]], dtype=torch.float)}),
        (row['Official Symbol B'], {"x": torch.tensor([role_map["Target"]], dtype=torch.float)})
    ])
    PTM_graph.add_edge(row['Official Symbol A'],row['Official Symbol B'],Author=row['Author'],
                       Organism_A=row['Organism A'], Organism_B=row['Organism B'], Type=row['PTM'])
```

```python
for node in PTM_graph.nodes():
    embedding = embedding_dict.get(node)  # Lookup by symbol
    if embedding is not None:
        embedding_tensor = torch.tensor(embedding, dtype=torch.float)
    else:
        embedding_tensor = torch.zeros(1280)  # Assuming 1280-d ESM embedding

    role_tensor = PTM_graph.nodes[node]["x"]
    combined_features = torch.cat([role_tensor, embedding_tensor])

    PTM_graph.nodes[node]["x"] = combined_features
```

```python
import torch.nn.functional as F
from torch.nn import Linear, ReLU
from torch_geometric.nn import Sequential, GCNConv
num_classes = len(PTM_df['PTM'].unique())
# GCN model
class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        '''

        :param in_channels: Dimension of the input features
        :param hidden_channels: Dimension of the hidden layer
        :param out_channels: Number of classes in output layer
        '''
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        '''

        :param x: Input node features
        :param edge_index: Graph edge indices
        :return: Ouput node features with message passing
        '''
        x = self.conv1(x, edge_index).relu()
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        return x

class EdgeMLP(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        '''

        :param in_channels: Dimension of the input features - concatenated embeddings
        :param hidden_channels: Dimension of the hidden layer
        :param out_channels: Dimension of the output layer
        '''
        super().__init__()
        self.layer1 = Linear(in_channels, hidden_channels)
        self.layer2 = Linear(hidden_channels, out_channels)
    def forward(self, x: torch.Tensor) -> torch.Tensor:
            x = F.relu(self.layer1(x))
            x = self.layer2(x)
            return x
GCN_model = GCN(
    in_channels=PTM_graph_py.x.shape[1],
    hidden_channels=128,
    out_channels=num_classes
)
edge_mlp = EdgeMLP(in_channels=2 * num_classes, hidden_channels=64, out_channels=num_classes)
```

```python
def build_edge_features(node_embeddings, edge_index):
    # node_embeddings: Tensor of shape [num_nodes, node_embedding_dim]
    # edge_index: Tensor of shape [2, num_edges]

    src = edge_index[0]
    tgt = edge_index[1]
    edge_features = torch.cat([
        node_embeddings[src],
        node_embeddings[tgt]
    ], dim=1)  # final shape: [num_edges, node_embedding_dim * 2]

    return edge_features
```

```python
# PyTorch Training loop
from sklearn.metrics import accuracy_score

optimizer = torch.optim.Adam(list(GCN_model.parameters()) + list(edge_mlp.parameters()), lr=0.01, weight_decay=5e-4)
for epoch in range(100):
    GCN_model.train()
    edge_mlp.train()
    optimizer.zero_grad() # Reset gradients to 0

    node_embeddings = GCN_model(PTM_graph_py.x, train_edge_index) # shape = [num_nodes, 128]
    train_edge_features = build_edge_features(node_embeddings, train_edge_index)
    train_logits = edge_mlp(train_edge_features)  # shape: [num_edges, num_classes]

    loss = F.cross_entropy(train_logits, train_edge_label)
    loss.backward()
    optimizer.step()
    GCN_model.eval()
    edge_mlp.eval()
    with torch.no_grad():
        # Training accuracy
        train_preds = train_logits.argmax(dim=1)
        train_acc = accuracy_score(train_edge_label.cpu(), train_preds.cpu())

        # Validation accuracy

        val_edge_features = build_edge_features(node_embeddings, val_edge_index)

        val_logits = edge_mlp(val_edge_features)
        val_preds = val_logits.argmax(dim=1)
        val_acc = accuracy_score(val_edge_label.cpu(), val_preds.cpu())

    print(f"Epoch {epoch:03d} | Loss: {loss.item():.4f} | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
```

```python
GCN_model.eval()
edge_mlp.eval()
with torch.no_grad():
    test_features = build_edge_features(node_embeddings, test_edge_index)
    test_logits = edge_mlp(test_features)
    test_preds = test_logits.argmax(dim=1)
    test_acc = (test_preds == test_edge_label).float().mean().item()

print(f"Test Accuracy: {test_acc:.4f}")
```
